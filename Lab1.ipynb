1. Translate the given image (lena translate.png) by (tx = 3.75, ty = 4.3) pixels.


```python
import numpy as np
from matplotlib import pyplot as plt
import math

srcImage=plt.imread("lena_translate.png")                  #Read the image
plt.imshow(srcImage,cmap='gray')                           #Display the image
plt.title("Orginal image")
srcSize=srcImage.shape                                     #Find the size of the image
tx=3.75                                                    #Define X shift
ty=4.3                                                     #define Y shift
trgtImage=np.zeros(srcSize)                                #Declare target image o same size filled with zero

for i in range(srcSize[0]):                                
    for j in range(srcSize[1]):
        X=math.floor(i-ty)                                 #find X' and Y'
        Y=math.floor(j-tx)
        a=(i-ty)-X                                         #Find the deviation from actual point and X' Y'
        b=(j-tx)-Y
        if(0<X<srcSize[0] and 0<Y<srcSize[1]):             #Applying bilinear nterpolation
            trgtImage[i,j]=(1-a)*(1-b)*srcImage[X,Y]+(1-a)*(b)*srcImage[X,Y+1]+(a)*(1-b)*srcImage[X+1,Y]+(a)*(b)*srcImage[X+1,Y+1]
plt.figure()                                               #Display output
plt.imshow(trgtImage,cmap='gray')
plt.title("Translated image")

```

Observation:Translated the image by tx = 3.75, ty = 4.3 pixels.Obtained a little bit smoothened image.

2. Rotate the given image (pisa rotate.png) about the image centre, so as to straighten the
   Pisa tower.

```python
import numpy as np
from matplotlib import pyplot as plt
import math
def rotate(angle,x,y):
    '''
     |newX|   |cos(angle)   sine(angle)| | x |
     |    | = |                        | |   |
     |newY|   |-sine(angle) cos(angle) | | y |
     
     '''
   
    cosine=math.cos(angle)                                 #Finding the cosine
    sine=math.sin(angle)                                   #Finding the sine
    newX=x*cosine+y*sine                                   #Applying the transform
    newY=-x*sine+y*cosine
    return newX,newY

srcImage=plt.imread("pisa_rotate.png")                     #Read the image
plt.imshow(srcImage,cmap='gray')                           #Display the image
plt.title("Orginal image")
srcSize=srcImage.shape                                     #Finding the size of the image
angle=4;                                                   #Defining the angle of rotation
angle=math.radians(angle)                                  #Converting Degree to radian
cosine=math.cos(angle)                                     #Finding the cosine
sine=math.sin(angle)                                       #Finding the sine
trgtImage=np.zeros(srcSize)                                #Declaring the target image
centre_height   = round(((srcImage.shape[0]+1)/2)-1)       # Find the centre of the image
centre_width    = round(((srcImage.shape[1]+1)/2)-1)

for i in range(srcSize[0]):
    for j in range(srcSize[1]):
        x=srcSize[0]-1-i-centre_height                     #Co-ordinates of pixel with respect to the centre   
        y=srcSize[1]-1-j-centre_width 
        new_x,new_y=rotate(angle,x,y)                      #Applying the transform
        new_x=centre_height-new_x                          #Change new_x and new_y with respect to the new centre
        new_y=centre_width-new_y
        X=math.floor(new_x)                                #Find X' Y'
        Y=math.floor(new_y)
        a=new_x-X                                          #Finding the deviation from actual point and X' Y'
        b=new_y-Y
        if(0<X+1<srcSize[0] and 0<Y+1<srcSize[1]):         #Applying bilinear interpolation
            trgtImage[i,j]=(1-a)*(1-b)*srcImage[X,Y]+(1-a)*(b)*srcImage[X,Y+1]+(a)*(1-b)*srcImage[X+1,Y]+(a)*(b)*srcImage[X+1,Y+1]

plt.figure()
plt.imshow(trgtImage,cmap='gray')                          #Display Rotated image
plt.title("Rotated image")
```

Observation:Rotated the given image about the image centre and found out the angle as 4 degree .

3. Scale the given image (cells scale.png) by 0.8 and 1.3 factors.

```python
import numpy as np
from matplotlib import pyplot as plt
import math

srcImage=plt.imread("cells_scale.png",0)                          #Read the image                 
plt.imshow(srcImage,cmap='gray')                                  #Display the image
plt.title("Orginal image")
srcSize=srcImage.shape                                            #Finding the size of the image
Xscale=0.8                                                        #Defining the X scaling
Yscale=0.8                                                        #Defining the Y scaling
scaledSize=[int(srcSize[0]*Yscale),int(srcSize[1]*Xscale)]        #Define target image dimension
trgtImage=np.zeros(srcSize)                                       #Define target image of new dimensions filled with zero
for i in range(scaledSize[0]): 
    for j in range(scaledSize[1]):
        X=math.floor(i/Yscale)                                    #Find X' Y'
        Y=math.floor(j/Xscale)
        a=(i/Yscale)-X                                            #Finding the deviation from actual point and X' Y'
        b=(j/Xscale)-Y
        if(0<X+1<srcSize[0] and 0<Y+1<srcSize[1]):                #Applying bilinear interpolation
            trgtImage[i,j]=(1-a)*(1-b)*srcImage[X,Y]+(1-a)*(b)*srcImage[X,Y+1]+(a)*(1-b)*srcImage[X+1,Y]+(a)*(b)*srcImage[X+1,Y+1]
plt.figure()
plt.imshow(trgtImage,cmap='gray')                                 #Display scaled image
plt.title("Scaled by 0.8")
Xscale=1.3                                                        #Defining the X scaling
Yscale=1.3                                                        #Defining the Y scaling
scaledSize=[int(srcSize[0]*Yscale),int(srcSize[1]*Xscale)]        #Define target image dimension
trgtImage=np.zeros(srcSize)                                       #Define target image of new dimensions filled with zero
for i in range(srcSize[0]):
    for j in range(srcSize[1]):
        X=math.floor(i/Yscale)                                    #Find X' Y'
        Y=math.floor(j/Xscale)
        a=(i/Yscale)-X                                            #Finding the deviation from actual point and X' Y'
        b=(j/Xscale)-Y
        if(0<X+1<srcSize[0] and 0<Y+1<srcSize[1]):                #Applying bilinear interpolation
            trgtImage[i,j]=(1-a)*(1-b)*srcImage[X,Y]+(1-a)*(b)*srcImage[X,Y+1]+(a)*(1-b)*srcImage[X+1,Y]+(a)*(b)*srcImage[X+1,Y+1]
plt.figure()
plt.imshow(trgtImage,cmap='gray')                                 #Display scaled image
plt.title("Scaled by 1.3")
```

Observation:Scaled the given image by 0.8 and 1.3 factors.Obtained zoomed in image for scaling factor=1.3 and zoomed out for scaling factor=0.8.
